### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt 
import numpy as np


x=np.arange(1, 6) 
y=x ** 2 
categories= ['A', 'B', 'C', 'D' ]
values=[10, 15, 7, 12]
scatter_x=np.random.rand(20)
scatter_y=np.random.rand(20)
pie_labels=['Python', 'Java', 'C++', 'JavaScript' ]
pie_sizes=[40, 25, 20, 15]


fig, axs= plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y, marker='o', color='b')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')


axs[0, 1].bar(categories, values, color='skyblue')
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

#3. Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color='r') 
axs[1, 0].set_title('Scatter Plot') 
axs[1, 0].set_xlabel('X-axis') 
axs[1, 0].set_ylabel('Y-axis')

#4. fie chart


axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90,colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
axs[1, 1].set_title('Pie Chart')


plt.tight_layout()


plt.show()
```
